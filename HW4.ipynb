## 1.2. Изучите стоимости недвижимости

```python
plt.figure(figsize=(8,6))
plt.hist(df["price"], bins = 50)
plt.title("Распределение недвижимости по его стоимости")
plt.xlabel("Стоимость недвижимости, д.е.")
plt.ylabel("Количество недвижимости, шт");
```

## 1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize = (6,4))
sns.histplot(df['sqft_living'], bins = 50)
plt.xlabel('Жилые квадратные метры')
plt.ylabel('Количество недвижимости, шт')
plt.title('Распределение недвижимости по их жилым кв. м');
```

## 1.4. Изучите распределение года постройки

```python
data_yr = df['yr_built'].value_counts(ascending=True)
```

```python
years_list = [1900+i*10 for i in range (0,13)]
```

```python
names = data_yr.index
values = data_yr.values
yr_list = list(range(1900,2025,5))
plt.figure(figsize=(16, 6))
plt.bar(names,values, width=0.7, facecolor='c', alpha=0.75)
plt.title('Распределение строительства недвижимости по годам')
plt.xlabel('Год постройки')
plt.ylabel('Количество недвижимости, шт.')
plt.grid(True)
plt.xticks(yr_list,rotation=40);
plt.show()
```



Как видно из графика, количество строиющейся недвижимости имеет положительную динамику роста. Однако, отмечается падение темпов строительства недвижимости в 1934-1935 годы и в 1945 году. В 1972 и в 1982 годах строительство недвижимости сократилость практически до уровня 1945 года.  Пик строительста недвижимости приходится на 1968, 1977, 2006 и 2014 годы.

## 2.3. Изучите распределение состояния домов

```python
con = df['condition'].value_counts()
names = con.index
values = con.values
```

# 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости на основе матрицы корреляции.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize = (6,6))
sns.heatmap(corr_matrix);
```

Проанализировав информация с графиков, наибольшее влияние на стоимость недвижимости имеет такие характеристики как: общая площадь, жилая площадь, общее состояние, количество ванных, спален, год постройки, этаж.
